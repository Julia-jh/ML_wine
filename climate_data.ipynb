```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("./Climate/climate_data_2020-2030.csv")
```

```python
df.info()
```

```python
df.set_index(['Unnamed: 0'])
df.head()
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
df.head()
```

```python
df['time'] = pd.to_datetime(df['time'])
df['time_bounds'] = pd.to_datetime(df['time_bounds'])
```

```python
# 북반구/남반구를 나누어주는 기준을 확인합니다. 1: Northern / 0: Southern 
df['Northern'] = np.where(df['lat']>0, 1, 0)
df.head()
```

```python
len(df['lat'].unique())
```

```python
len(df['lon'].unique())
```

```python
# 2021 data 추출하기
df_2021 = df[df['time_bounds'].dt.year == 2021].copy()
df_2021.info()
```

```python
df_2021.drop(['time'], axis = 1, inplace = True)
df_2021.head()
```

```python
df_2021['year_month'] = df_2021['time_bounds'].dt.to_period('M')
df_2021.head()
```

```python
df_2021 = df_2021.groupby(['lat', 'lon', 'year_month']).agg({
    'TG': 'mean',
    'TN': 'min',
    'TX': 'max',
    'RR': 'mean',
    'RR1': 'mean'
}).reset_index()

df_2021
```

```python
# 중간 저장하기
df_2021.to_csv('./Climate/cilmate_data_2021.csv')
```

```python
df_2021['Northern'] = np.where(df_2021['lat']>=0, 1, 0)
df_2021.head()
```

북반구/남반구별로 포도 성장 시기가 다릅니다. 따라서, 북반구와 남반구로 데이터프레임을 나누어 각 시기에 맞게 성장시기의 TG, TN, TX, RR, RR1 값을 구합니다.
- TG, RR, RR1은 평균 수치로 그대로 평균으로 하면 될 것 같은데...
- TN: 성장시기동안의 최저 기온으로 하겠음
- TX: 성장시기동안의 최고 기온으로 하겠음

포도 성장 시기
- 북반구: 4월 ~ 10월
- 남반구: 10월 ~ 4월

```python
df_2021_north = df_2021[df_2021['Northern'] == 1]
df_2021_south = df_2021[df_2021['Northern'] == 0]

df_2021_north['growing_season'] = np.where((df_2021_north['year_month'].dt.month>= 4)&(df_2021_north['year_month'].dt.month <= 10),1,0)
```

```python
df_2021_south['growing_season'] = np.where((df_2021_south['year_month'].dt.month>= 10)|(df_2021_south['year_month'].dt.month <= 4),1,0)
```

```python
df_2021_north.head()
```

```python
df_2021_north = df_2021_north.groupby(['lat', 'lon', 'growing_season']).agg({
    'TG': 'mean',
    'TN': 'min',
    'TX': 'max',
    'RR': 'mean',
    'RR1':'mean'
}).reset_index()

df_2021_north = df_2021_north[df_2021_north['growing_season'] == 1]
df_2021_north
```

```python
df_2021_south.head(12)
```

```python
df_2021_south = df_2021_south.groupby(['lat', 'lon', 'growing_season']).agg({
    'TG': 'mean',
    'TN': 'min',
    'TX': 'max',
    'RR': 'mean',
    'RR1':'mean'
}).reset_index()

df_2021_south
```

```python
df_2021_south = df_2021_south[df_2021_south['growing_season'] == 1]
df_2021_south
```

```python
df_2021 = pd.concat([df_2021_north, df_2021_south], axis = 0)
df_2021.head()
```

```python
df_2021.drop(['growing_season'], axis = 1, inplace= True)
```

```python
df_2021.reset_index(inplace= True)
```

```python
df_2021.drop('index', axis = 1, inplace= True)
```

```python
df_2021.info()
```

```python
df_2021.to_csv('./Climate/climate_data_2021_growing_season.csv')
```
