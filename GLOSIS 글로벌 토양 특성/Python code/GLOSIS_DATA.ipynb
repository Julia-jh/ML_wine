### 컬럼 정보

- ID : 데이터 베이스 id
- HWSD2_SMU_ID ,HWSD1_SMU_ID ,WISE30s_SMU_ID: Soil Mapping Unit 번호
- COVERAGE : Data source: ESDB, SOTWIS, DSMW, China, Afghanistan, Ghana, Türkiye.
- SHARE(Share of the soil unit) : share of the soil unit within the mapping unit in percentage.
-  WRB4 : the name of the Soil Reference Group plus the most important
primary qualifier(4글자)
- WRB_PHASES :  the full symbol of the WRB soil name taking
into account phases and referring to the Soil Reference Group plus all primary qualifier
- WRB2 : the name of the Soil Reference Group plus the most important
primary qualifier(2글자)
- WRB2_CODE : the name of the Soil Reference Group plus the most important
primary qualifier(코드번호)
- FAO90 :  the name of the soil unit in the FAO90 Revised Legend
- KOPPEN :  Kӧppen-Geiger climate classes (Tropical, Arid, Temperate,
Cold and Polar) 
- TEXTURE_USDA : Texture class (USDA conventions)
- REF_BULK_DENSITY : Reference bulk density is a property of particulate
materials. It is the mass of many particles of the material divided by the volume they occupy.
The volume includes the space between particles as well as the space inside the pores of
individual particles
- BULK_DENSITY : Bulk density is defined as “the mass of the many particles of the
material divided by the total volume they occupy”
- DRAINAGE : soil drainage refers to the natural capability of a soil to
remove excess water.

- ROOT_DEPTH : the depth to which plants can exploit the soil for nutrients and
moisture.

- AWC :  the maximum amount of water that the soil can hold that is
available for plant growth

- PHASE1 ,PHASE2 : phases are subdivisions of soil units based on characteristics which
are significant for the use or management of the land but are not diagnostic for the
separation of the soil units themselves

- ROOTS : Obstacle to roots: provides the depth class of an obstacle to roots.(ESDB)
- IL : Impermeable layer: indicates the presence of an impermeable layer within the soil
profile. (ESDB)

- ADD_PROP (Additional property) :  certain soil properties, inherent to the soil unit definition
that are relevant for agricultural use of the soil are vertic and gelic. The additional field provides
details on gelic properties (gelic soil groups in FAO90) and vertic properties (Vertisols, vertic
soil groups in FAO90).

- SWR : Soil Water Regime (ESDB)

- LAYER : Depth Layer (from D1 to D7)

- TOPDEP : Depth of top of layer

- BOTDEP : Depth of bottom of layer

- COARSE : Coarse fragments
- SAND 
- SILT 
- CLAY
- TEXTURE_SOTER : Texture class (SOTER conventions)
- BULK : Bulk Density
- REF_BULK : Reference Bulk Density
- ORG_CARBON : Organic Carbon Content
- PH_WATER :pH in water
- TOTAL_N :Total nitrogen content
- CN_RATIO :Carbon/Nitrogen ratio (C/N)
- CEC_SOIL : CEC soil, CATION EXCHANGE CAPACITY (CEC) of the fine earth fraction (cmolc/kg). The total nutrient
fixing capacity of a soil is well expressed by its Cation Exchange Capacity. 
- CEC_CLAY :CEC clay,corrected for contribution of organic matter (cmolc/kg). 
- CEC_EFF :. ECEC is defined as the sum of exchangeable (Ca++, Mg++,
K+, Na+) plus the sum of exchangeable (H+, Al+++) (Van Reeuwijk, 2002).
- TEB :Total exchangeable bases stand for the sum of
exchangeable cations in a soil: sodium (Na+), calcium (Ca++), magnesium (Mg++) and Potassium
(K+).
- BSAT : The base saturation measures the sum
of exchangeable cations (nutrients) Na+, Ca++, Mg++ and K+ as a percentage of the overall
exchange capacity of the soil (including the same cations plus H+ and Al+++)
- ALUM_SAT : The exchangeable aluminum
percentage (ALSA) has been set at zero when pHwater is higher than 5.5
-  ESP : The exchangeable sodium percentage has been used to indicate levels of sodium in soils.
- TCARBON_EQ : Calcium Carbonate
- GYPSUM : Gypsum is a chemical compound (a salt) which
occurs occasionally in soils particularly in the driest areas of the globe where it can occur in a
flower-like form typically opaque with embedded sand grains called desert rose. 
- ELEC_COND :  Coastal and desert soils can be enriched with water-soluble
salts or salts more soluble than gypsum. The salt content of a soil can be roughly estimated from
the Electrical Conductivity of the soil measured in a saturated soil paste or a more diluted
suspension of soil in water

19. COARSE: 굵은 입자 비율.
20. SAND: 모래 비율.
21. SILT: 미사 비율.
22. CLAY: 점토 비율.

24. BULK: 상대 밀도.
25. ORG_CARBON: 유기 탄소 함량.

27. TOTAL_N: 총 질소 함량.
28. CN_RATIO: 탄소/질소 비율.

31. CEC_CLAY: 점토의 양이온 교환 능력.

```python
import geopandas as gpd
from geopy.geocoders import GoogleV3
from shapely.geometry import Point

# Google Geocoding 서비스 초기화
API_KEY = ''
geolocator = GoogleV3(api_key=API_KEY)

# GeoDataFrame에서 첫 번째 좌표 가져오기 (예제)
geo_country = df.copy()

def get_country_region(geometry):
    if geometry.geom_type == 'LineString':
        point = geometry.interpolate(0.5)  # 중앙 지점을 사용
    elif geometry.geom_type == 'MultiLineString':
        point = geometry.interpolate(0.5)  # 첫 번째 LineString 사용

    location = geolocator.reverse((point.y, point.x))
    if location:
        address = location.raw.get('address_components', [])
        country = state = None
        for component in address:
            if 'country' in component['types']:
                country = component['long_name']
            elif 'administrative_area_level_1' in component['types']:
                state = component['long_name']
        return country, state

# 각 지오메트리의 나라와 지역 정보 가져오기
geo_country['country_region'] = geo_country['geometry'].apply(get_country_region)
geo_country

```
