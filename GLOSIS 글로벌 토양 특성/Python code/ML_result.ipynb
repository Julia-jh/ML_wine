```python
import pandas as pd

pred_df = pd.read_csv('Wine_ML_predict_result_2021-2070.csv',index_col=0)
pred_df
```

```python
locat = pd.read_csv('../Project_data/wine_total_data_0721.csv',index_col=0)
locat
```

```python
locat['Grape_Variety'] = locat['Grape_Variety'].str.strip()
len(locat['Grape_Variety'].unique())
```

```python
locat.loc[locat['Grape_Variety'].isin(['albarin negro mencia']), 'Grape_Variety'] = 'albarin negro'
locat.loc[locat['Grape_Variety'].isin(['aragonez', 'tinta roriz', ]), 'Grape_Variety'] = 'tempranillo'
locat.loc[locat['Grape_Variety'].isin(['cabernet', 'cabernet savignon']), 'Grape_Variety'] = 'cabernet sauvignon'
locat.loc[locat['Grape_Variety'].isin(['carinena']), 'Grape_Variety'] = 'carignan'
locat.loc[locat['Grape_Variety'].isin(['garnacha', 'grenache tinta', ]), 'Grape_Variety'] = 'grenache'
locat.loc[locat['Grape_Variety'].isin(['garnacha tintorera', 'grenache tintorera']), 'Grape_Variety'] = 'alicante bouschet'
locat.loc[locat['Grape_Variety'].isin(['pinot noir precoce']), 'Grape_Variety'] = 'pinot noir'
locat.loc[locat['Grape_Variety'].isin(['shiraz and syrah']), 'Grape_Variety'] = 'syrah'
locat.loc[locat['Grape_Variety'].isin(['tanat']), 'Grape_Variety'] = 'tannat'

print(locat['Grape_Variety'].nunique(), locat['Grape_Variety'].unique())
```

```python
pred_df.loc[pred_df['Year'] == 2035]
```

```python
pred_df['Year'].unique()
```

```python
locat.reset_index(drop=True,inplace=True)
```

```python
#locat을 옆으로 이어붙이기

year= [2021, 2025, 2030, 2035, 2040, 2045, 2050, 2055, 2060, 2065, 2070]

total_df = pd.DataFrame()

for n in year:

    #pred_df에서 년도별로 추출
    temp = pred_df.loc[pred_df['Year'] == n].copy()
    temp.reset_index(drop=True,inplace=True)

    location = locat[['Longitude','Latitude','Continent','Country','Region Lv1']]

    temp = temp.merge(location, left_index=True, right_index=True, how='left')


    if total_df.empty:
        total_df = temp
    else:
        total_df = pd.concat([total_df, temp], axis=0)


total_df



```

```python
# 지역별 예측 그래프 함수

def draw_region_pred(region,model='y_pred_tree'):
    temp = total_df.loc[total_df['Region Lv1'] == region].groupby(['Year',model])[model].count().to_frame()
    temp.columns = ['Count']
    temp.reset_index(inplace=True)
    temp[model] = temp[model].map(idx_grape)
    temp.loc[temp[model].isnull(),model] = 'other'

    fig, ax = plt.subplots(1, 2, figsize=(17, 6))

    # 2021년 데이터를 Count 기준으로 내림차순 정렬
    if model == 'y_pred_tree':
        color = 'orange'
    else: 
        color = 'pink'
    
    data_2021 = temp.loc[temp.Year == 2021].sort_values(by='Count', ascending=False)
    ax[0].bar(data=data_2021, x= model, height='Count')
    ax[0].tick_params(axis='x', rotation=45)
    ax[0].set_title('2021년 {} 지역 재배 품종'.format(region))

    # 2070년 데이터를 Count 기준으로 내림차순 정렬
    data_2070 = temp.loc[temp.Year == 2070].sort_values(by='Count', ascending=False)
    ax[1].bar(data=data_2070, x=model, height='Count',color=color)
    ax[1].tick_params(axis='x', rotation=45)
    ax[1].set_title('2070년 {} 지역 재배 품종 ({} 예측)'.format(region,model))

    plt.show()

    
    
```

```python
def draw_region_pred_year(region, year1, year2, model='Voted_pred'):
    temp = total_df.loc[total_df['Region Lv1'] == region].groupby(['Year',model])[model].count().to_frame()
    temp.columns = ['Count']
    temp.reset_index(inplace=True)
    temp[model] = temp[model].map(idx_grape)
    temp.loc[temp[model].isnull(),model] = 'other'

    fig, ax = plt.subplots(1, 2, figsize=(17, 6))

    # 2021년 데이터를 Count 기준으로 내림차순 정렬
    
    data_year1 = temp.loc[temp.Year == year1].sort_values(by='Count', ascending=False)
    ax[0].bar(data=data_year1, x= model, height='Count',color='orange')
    ax[0].tick_params(axis='x', rotation=45)
    ax[0].set_title('{}년 {} 지역 재배 품종 ({} 예측)'.format(year1,region,model))

    # 2070년 데이터를 Count 기준으로 내림차순 정렬
    data_year2 = temp.loc[temp.Year == year2].sort_values(by='Count', ascending=False)
    ax[1].bar(data=data_year2, x=model, height='Count',color='purple')
    ax[1].tick_params(axis='x', rotation=45)
    ax[1].set_title('{}년 {} 지역 재배 품종 ({} 예측)'.format(year2,region,model))

    plt.show()
```
