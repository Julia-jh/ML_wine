# 와인생산지별 토양 특성 EDA

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

sns.countplot(x=df['KOPPEN'])
plt.show()
```

- A	Tropical(열대)
- B	Arid(건조)
- C	Temperate(온대),
- D	Cold(대륙)
- E	Polar(극지)

> 와인 생산지들의 Koppen 기준 기후 특성은 주로 Temperate(온대) 와 Cold(대륙)이다.

- 온대 기후는 가장 추운 달의 평균 기온이 0°C(32°F) [10] (또는 -3°C(27°F)) [7] 와 18°C(64.4°F) 사이이고, 적어도 한 달은 평균 기온이 10°C(50°F)를 넘습니다.

- 대륙성 기후는 적어도 한 달은 평균 0°C(32°F) 미만(또는 -3°C(27°F))이고 적어도 한 달은 평균 10°C(50°F) 이상입니다. 

```python
temp=df['WRB_PHASES_y'].value_counts().sort_values(ascending=False)[:10]
temp
```

```python
temp.values
```

```python
sns.barplot(x=temp.index, y=temp.values)
plt.xticks(rotation=45)
plt.show()
```

- 주요 토양 참조 그룹과 주요 1차 지시자를 나타냄

- TC : Technosol
- CMDy : Dystric Cambisol
- LPli : Lithic Leptosol

>TC(Technosol)는 기술적인 인간 활동에 의해 형성된 토양을 나타내는 용어입니다. 이 토양은 주로 도시화된 지역에서 발생하며, 건설 현장이나 산업 지역에서의 토양 개질이나 이동에 의해 형성됩니다.

>Cambisols는 토양의 발달 과정에서 Cambic 토양층을 가지고 있는 젊은 토양을 나타냅니다.Dystric Cambisols은 혼합 경작에 사용됩니다.주로 지중해 지역에서 발생하는 토양으로, 반습 기후 지역을 경계로 하는 semi-arid 기후 지역에서 널리 분포합니다.

>Lithic Leptosol은 연속된 암석 위의 매우 얕은 토양 또는 거친 파편( 자갈 및/또는 돌 )이 매우 풍부한 더 깊은 토양입니다. 토양은 배수 능력이 좋지만, 물을 오래 머금지 못하기 때문에 건조한 상태를 유지하기 쉽습니다.

- 이상하게 토양 그룹은 식물이 자라기 어려운 곳으로 많이 나온다...?
- 이거 아마 위, 경도가 지역 으로 되어 있어서 그런거 같다

# 지도 시각화 

```python
df_2 = pd.read_csv('../Project_data/wine_soil_data.csv')
df_2.head()
```

```python
#와인 생산지역 지도 시각화

import folium

my_map = folium.Map(location=[0,0],zoom_start=1)

for n in range(len(df_2)):
    folium.Marker(
        location=[df_2['Latitude'][n],df_2['Longitude'][n]],
        tooltip=df_2['Country'][n] + ' ' + df_2['Region Lv1'][n],
        icon=folium.Icon(icon='tree',icon_color='purple',prefix='fa')
    ).add_to(my_map)

my_map
```
